```python
import matplotlib.pyplot as plt
import numpy as np

speedup = [1, 2,4,4,4, ,6,6, 12,12]
accuracy_at_20 = [79.73, 73.88, 77.01, 76.09, 67.59]
accuracy_at_40 = [83.05, 79.45, 81.80,80.94,72.96]
accuracy_at_100 = [85.84,84.74,85.76,84.88,79.47]
accuracy_at_200 = [88.42, 87.26,87.76,87.01,82.71]
degradation_at_20 = [(accuracy_at_20[i]/accuracy_at_20[0]) for i in range(5)]
degradation_at_40 = [(accuracy_at_40[i]/accuracy_at_40[0]) for i in range(5)]
degradation_at_100 = [(accuracy_at_100[i]/accuracy_at_100[0]) for i in range(5)]
degradation_at_200 = [(accuracy_at_200[i]/accuracy_at_200[0]) for i in range(5)]
names = ['BERT Base (12 layers)', 'DistilBERT (6 layers)','BERT Base (6 layers)', 'Asymmetry KALE BERT Base (6 layers)','BERT Base (3 layers)', 'Asymmetry KALE BERT Base (3 layers)', 'BERT Base (2 layers)','Asymetry KALE BERT Base (2 layers)','Asymetry KALE BERT Base (1 layers)','BERT Base (1 layers)']
```

```python
degradation_at_20 
```

```python
order = np.argsort(degradation_at_20)
speedup_ord = [speedup[i] for i in order]
deg_ord = [degradation_at_20[i] for i in order]
name_ord = [names[i] for i in order]

plt.figure(figsize=(12,12))
plt.plot(speedup_ord, deg_ord, 'o')
plt.xlabel('speedup')
plt.ylabel('degradation')
plt.grid()

for s, d, n in zip(speedup_ord, deg_ord, name_ord):
    plt.annotate(n, # this is the text
                     (s,d), # these are the coordinates to position the label
                     textcoords="offset points", # how to position the text
                     xytext=(0,10), # distance from text to points (x,y)
                     ha='center') # horizontal alignment can be left, right or center
    
plt.savefig('KALE-NQ-20.png')
```
